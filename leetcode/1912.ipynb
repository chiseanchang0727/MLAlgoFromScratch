```python
movie_set = [
    [0, 1, 5],
    [0, 2, 6],
    [0, 3, 7],
    [1, 1, 4],
    [1, 2, 7],
    [2, 1, 5]
]
```

```python
from collections import defaultdict

movie_dict = defaultdict(list)

for movie in movie_set:
    movie_dict[movie[1]].append([movie[0], movie[2], False])

# sort each list by movie[2] descending
for key in movie_dict:
    movie_dict[key].sort(key=lambda x: (x[1], x[0]), reverse=False)

print(dict(movie_dict))
```
