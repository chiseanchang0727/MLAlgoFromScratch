```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def print_list(node):
    vals = []
    while node:
        vals.append(str(node.val))
        node = node.next
    return " → ".join(vals) if vals else "None"
```

Reversed list

```python
# --- Build sample list: 1 → 2 → 3 ---
head = ListNode(1, ListNode(2, ListNode(3)))

print("Inital: ")
print(f"reversed: prev = {print_list(None)}")
print(f"unreversed: curr = {print_list(head)}")
print("=" * 40)

# --- Reversal with tracing ---
prev = None
curr = head

while curr:
    temp = curr.next      # 1) 暫存下一個
    curr.next = prev      # 2) 反轉
    prev = curr           # 3) 前進 prev
    curr = temp           # 4) 前進 curr

    print(f"reversed: prev = {print_list(prev)}")
    print(f"unreversed: curr = {print_list(curr)}")
    print("=" * 40)

print("Final：")
print(f"head = {print_list(prev)}")
```

Note that:
- `prev` = head of the reversed portion
- `curr` = head of the unreversed portion

```python
list1 = ListNode(1, ListNode(3, ListNode(5, ListNode(7))))
list2 = ListNode(2, ListNode(4, ListNode(6)))
```

```python
print_list(list1)
```

```python
print_list(list2)
```

```python
dummy = ListNode() # create the merged list
tail = dummy # define head
```

```python
while list1 and list2:
    if list1.val <= list2.val:
        tail.next = list1 # attach the current list_pointer to the merged list
        list1 = list1.next
        tail = tail.next # move forward by one
    else:
        tail.next = list2        
        list2 = list2.next
        tail = tail.next # move forward by one

if list1 or list2:
    tail.next = list1 if list1 else list2
```
