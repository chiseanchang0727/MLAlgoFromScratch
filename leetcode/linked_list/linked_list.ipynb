Reversed list

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def print_list(node):
    vals = []
    while node:
        vals.append(str(node.val))
        node = node.next
    return " → ".join(vals) if vals else "None"

# --- Build sample list: 1 → 2 → 3 ---
head = ListNode(1, ListNode(2, ListNode(3)))

print("Inital: ")
print(f"reversed: prev = {print_list(None)}")
print(f"unreversed: curr = {print_list(head)}")
print("=" * 40)

# --- Reversal with tracing ---
prev = None
curr = head

while curr:
    temp = curr.next      # 1) 暫存下一個
    curr.next = prev      # 2) 反轉
    prev = curr           # 3) 前進 prev
    curr = temp           # 4) 前進 curr

    print(f"reversed: prev = {print_list(prev)}")
    print(f"unreversed: curr = {print_list(curr)}")
    print("=" * 40)

print("Final：")
print(f"head = {print_list(prev)}")
```

Note that:
- `prev` = head of the reversed portion
- `curr` = head of the unreversed portion
